```r
data <- read.csv("/Users/kosuke/thesis/airport_thesis/data/mid/corpus/recommend.csv", stringsAsFactors=F)
train_corpus <- read.csv("/Users/kosuke/thesis/airport_thesis/data/mid/corpus/train_corpus.csv")
test_corpus <- read.csv("/Users/kosuke/thesis/airport_thesis/data/mid/corpus/test_corpus.csv")

bound <- min(test_corpus[,2])
test_y <- data[bound:nrow(data),]
test_y <- as.vector(sapply(test_y, replace_y, "no"))

train_corpus <- as.matrix(train_corpus)
train_corpus <- cbind(train_corpus, 1)


set.seed(123)
x <- rbinom(length(test_y),1, sum(test_y)/length(test_y))
tag <- data.frame(doc_index=seq(max(train_corpus[,2]) + 1, 7437), tag_index=x)
tag_random <- inner_join(test_corpus, tag, by="doc_index")[,5]
test_corpus[,4] <- tag_random
test_corpus <- as.matrix(cbind(test_corpus, 0))

corpus <- rbind(train_corpus, test_corpus)

V <- max(corpus[,1])
D <- max(corpus[,2])
```

```r
t <- proc.time()
k <- c(seq(10, 100, 10), 150, 200, 300)
it = 0
alpha   <- 0.1
beta    <- 0.1

LL <- as.vector(NULL)
WL <- as.vector(NULL)
YL <- as.vector(NULL)
burn_in = 1000
total = 2000
res <- list()
for(i in k){
    gamma   <- 0.1 * nrow(corpus)/i
    it = it + 1
    set.seed(123)
    res[[it]] <- LDAcgs2(corpus, i, alpha, beta, gamma, total, burn_in, 1, 0, 0)
    LL[it] <- burn_in / (sum(1 / res[[it]]$likelihood[(burn_in + 1):total]))
}
proc.time() - t
```

## Accuracy

```r
y <- apply(res[[11]]$y[(nrow(train_corpus)+1):nrow(corpus), ], 1, function(x){
    return (which(x == max(x)) - 1)
})
result <- cbind(test_corpus, unlist(y))
y_pred <- NULL
y_prop_deter <- NULL
for(d in (bound):max(test_corpus[,2])){
    res_d <- result[result[,2] == d, ]
    y_prop_deter[d - (bound - 1)] <- mean(res_d[, 6])
    if(mean(res_d[, 6]) <= 0.5){
        y_pred[d - (bound - 1)] <- 0
    }
    else{
        y_pred[d - (bound - 1)] <- 1
    }
}
accuracy_deter <- calc_accuracy(y_pred, test_y)
```

```r
y <- apply(res[[10]]$y[(nrow(train_corpus)+1):nrow(corpus), ], 1, function(x){
    return (which(x == max(x)) - 1)
})
result <- cbind(test_corpus, unlist(y))
y_pred <- NULL
y_prop_deter <- NULL
for(d in (bound):max(test_corpus[,2])){
    res_d <- result[result[,2] == d, ]
    y_prop_deter[d - (bound - 1)] <- mean(res_d[, 6])
    if(mean(res_d[, 6]) <= 0.5){
        y_pred[d - (bound - 1)] <- 0
    }
    else{
        y_pred[d - (bound - 1)] <- 1
    }
}
accuracy_deter <- calc_accuracy(y_pred, test_y)
```

```r
pred_y <- res[[10]]$y[(nrow(train_corpus)+1):nrow(corpus), ]
y_prop <- apply(data.frame(pred_y), 1, function(x){
    return (x[2] / 1000)
})

result <- cbind(test_corpus, y_prop)
y_pred <- NULL
y_prop <- NULL
for(d in (bound):max(test_corpus[,2])){
    res_d <- result[result[,2] == d, ]
    y_prop[d - (bound - 1)] <- mean(res_d[, 6])
    if(mean(res_d[, 6]) <= 0.5){
        y_pred[d - (bound - 1)] <- 0
    }
    else{
        y_pred[d - (bound - 1)] <- 1
    }
}
accuracy_prob <- calc_accuracy(y_pred, test_y)
```

```r
mpqa <- read.csv("/Users/kosuke/thesis/airport_thesis/data/mid/corpus/lexicon.csv")
mpqa_corpus <- inner_join(data.frame(test_corpus), mpqa, by="word_index")
pol_mpqa <- as.vector(NULL)
mpqa_prop <- as.vector(NULL)
for (d in 1:(D-bound + 1)){
    corpus_d <- mpqa_corpus[mpqa_corpus$doc_index == bound + d - 1, ]
    len_neg <- sum(corpus_d$polarity == "negative")
    mpqa_prop[d] <- 1 - len_neg / nrow(corpus_d)
    if (len_neg >= nrow(corpus_d) / 2){
        pol_mpqa[d] <- 0
    }
    else{
        pol_mpqa[d] <- 1
    }
}
```

## polarity dictionary

```r
est_p_v <- function(phi, psi){
    numerator <- phi %*% psi
    denominator <- apply(phi, 1, sum)
    return (numerator / denominator)
}
```

```r
p_v <- data.frame(word_index=1:V, p_v=est_p_v(res[[10]]$phi, res[[10]]$psi))
p_v$polarity <- 0
p_v[p_v$p_v > 0.5, ]$polarity <- "positive"
p_v[p_v$p_v <= 0.5, ]$polarity <- "negative"
```

```r
vocab_mat <- inner_join(mpqa, p_v, by="word_index")
sum(table(vocab_mat$polarity.x, vocab_mat$polarity.y)[c(1,4)]) / nrow(vocab_mat)

pred     <- prediction(vocab_mat$p_v, vocab_mat$polarity.x)
auc.temp <- performance(pred, "auc")
auc   <- unlist(auc.temp@y.values)
auc
```
